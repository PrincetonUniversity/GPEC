```python
ax2.plot(x,ntv.interp(x,'T_phi')['T_phi'],ls= '',marker='o',
         label='Interpolation')
ax2.legend()
allfig #not necessary if your gui is still open
```

Note that it initialized the interpolation function on the first call,
and saved it internally for the second. Interpolation (like plotting)
can be done for any number of y values at once.

## The gpec Module

This module is for more advanced operators who want to control/run 
the fortran package for DCON, IPEC, and NAT3. To get started, be sure
to explore the package using the tab interface, '?' and other tools
mentioned obove. Many of the module's attributes are simply standard python 
modules imported by gpec.

Moving straight to the point, lets submit our own full IPEC run in just 3 lines:

```python
# steal the settings from a previous run (leaving run director as deafult)
new_inputs = gpec.InputSet(indir='/p/gpec/users/nlogan/data/example/').infiles
```

```python
# change some inputs
new_inputs['dcon']['DCON_CONTROL']['nn']=2
```

```python
# submit job to cluster
#WARNING: This clears files from the specified loc if it already exists (creates it otherwise)
gpec.run(loc='~/pypec_example/',**new_inputs)
```


Running the package using this module will 
1. Create the new location (if not already existing), and cd to it

2. Remove all .out, .dat, and .bin files already in the directory
      - not dcon ones if rundcon=False, not nat3 ones for runnat3=False, etc.

3. Copy all .dat and .in files from the run directoy to the new location
4. Write all namelists supplied in **kwargs to file (overwriting)
5.  - Write a unique bash script for the run, and submit it or 
    - Run each fortran routine from the commandline
6. Remove all .dat and xdraw files from the directory

You will note that there is significant oportunity for overwriting files,
and the user must be responsible for double checking that all inputs 
are correct and will not result in life crushing disaster for themselves
or a friend when that super special file gets over-written.

### Including Post Processing Data

Lets look through a huge file of first order perterbations
we want to define some new data from the raw outputs, plot it,
and save the results in a table

```python
pmodb, = data.read('ipec_pmodb_fun_n1.out')
pmodb.y['theta_cyl'] = arctan2(pmodb.y['z'],pmodb.y['r']-pmodb.params['R0'])
```

Thats right, we just added data!

```python
pmodb.y['kxprp'] = pmodb.y['Bkxprp']/pmodb.y['equilb']
modfig=pmodb.plot1d(['theta_cyl','kxprp'],'theta',x2rng=0.16)
modfig.printlines('mynewdataslice.out',squeeze=True)
```

Thats it! Read the table with your favorit editor. 
Those familiar with matplotlib will recognize that the Figure
base class was modified to include this ability (we warned you in the begining).
The file will probably need a little cleaning at the top since it

### 2D Displays

Lets look at some 2d data

```python
pb2, = data.read('ipec_pbrzphi_n1.out')
fig2d = pb2.plot2d('b_r')
```

Hm... not very pretty. Plus, we might be getting missing some structure. 
Are there any details in the core we are missing due to the large edge values?

```python
fig2d = pb2.plot2d('b_r',vmin=-0.6e-4,vmax=0.6e-4)
```

Looks good in the core. If you are thinking "But those weren't listed as keyword arguments!" then you have
been glossing over the **kwargs documentation. Most all pypec calls accept
key word arguments for their root functions (in this case, pyplot.pcolormesh).

Finally, we want to make sure the aspect ratio is right and the white space is removed.

```python
fig2d = pb2.plot2d('b_r',aspect='equal')
fig2d.get_axes()[0].set_xlim(1.75,2.15)
fig2d
```

There. That looks better. A beatiful and informative 2D image!
